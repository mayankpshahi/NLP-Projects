# Spam Classification Using different Vectorizer and classifiers

```python
import pandas as pd
import os

senti = pd.read_csv("spam.csv",encoding='cp1252')
#separate the features and the output variable

X = senti['v2']
Y= senti['v1']
```

```python
X.shape
```

```python
Y.shape
```

```python
Y.unique()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, train_size=0.9, random_state=1111)
#Split into train test
```

# Using CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
print("Now with Countvectorizer")
# make a sparse document term matrix using count vectorizer
vect = CountVectorizer( stop_words="english")
# X_train_dtm is the document term matrix
X_train_dtm = vect.fit_transform(X_train)
print(type(X_train_dtm))
X_test_dtm = vect.transform(X_test)
```

```python
vect.vocabulary_
```

```python
X_train_dtm
```

```python
print(X_train_dtm)
```

```python
# fit the document term matrix and output variable in a Multinomial naive bayes model
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()

nb.fit(X_train_dtm, y_train)

# predict for test samples document term matrix
y_pred = nb.predict(X_test_dtm)
```

```python
# print accuracy score and the confusion matrix
from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
#With Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state = 1)
dtc.fit(X_train_dtm, y_train)
y_pred = dtc.predict(X_test_dtm)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
#With Random forest Classifier
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_jobs = -1, random_state = 1)
rfc.fit(X_train_dtm, y_train)
y_pred = rfc.predict(X_test_dtm)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
#With Gradient Boosting Classifier
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier(random_state = 1)
gbc.fit(X_train_dtm, y_train)
y_pred = gbc.predict(X_test_dtm)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
#With SVM Classifier
from sklearn.svm import SVC
svmm = SVC()
svmm.fit(X_train_dtm,y_train)
y_pred = svmm.predict(X_test_dtm)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion matrix:")
print(confusion_matrix(y_test, y_pred))
```
